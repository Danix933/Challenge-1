# Informe Final: Análisis Comparativo de Tiendas
## Introducción

El presente informe tiene como objetivo analizar el desempeño de cuatro tiendas con el fin de recomendar al Sr. Juan en cuál de ellas debería vender sus productos.

Para ello, se evaluaron los siguientes factores clave:

1. Ingresos totales por tienda

2. Categorías de productos más y menos vendidas

3. Calificación promedio de los clientes

4. Productos más y menos vendidos

5. Costo de envío promedio

A partir de estos indicadores, se busca identificar la tienda con mejor equilibrio entre rentabilidad, volumen de ventas, satisfacción del cliente y costos operativos.

##  Desarrollo del análisis
### Ingresos totales

Los ingresos obtenidos por cada tienda fueron:

Tienda 1: $1.150.880.400

Tienda 2: $1.116.343.500

Tienda 3: $1.098.019.600

Tienda 4: $1.038.375.700

La Tienda 1 presenta el mayor ingreso total, lo que indica un fuerte desempeño comercial.

## Categorías más y menos vendidas

En las cuatro tiendas se observa un patrón consistente:

### Categorías más vendidas:

Muebles

Electrónicos

Juguetes

Esto indica que existe una demanda sólida y estable en estos segmentos en todas las tiendas.

### Categorías menos vendidas, se repiten principalmente:

Instrumentos musicales

Libros

Artículos para el hogar

La Tienda 3 destaca en ventas de Muebles (499) y mantiene cifras altas también en Electrónicos (451), lo que demuestra un volumen competitivo.

## Calificaciones promedio

### Las calificaciones promedio de los clientes son:

Tienda 1: 3.98

Tienda 2: 4.04

Tienda 3: 4.05

Tienda 4: 4.00

La Tienda 3 obtiene la mejor valoración, lo que refleja mayor satisfacción del cliente. Aunque las diferencias son pequeñas, en mercados competitivos incluso centésimas pueden representar una ventaja reputacional importante.

## Costos de envío promedio

### Los costos promedio de envío fueron:

Tienda 1: $26.018,61

Tienda 2: $25.216,24

Tienda 3: $24.805,68

Tienda 4: $23.459,46

La Tienda 4 tiene el menor costo de envío, lo que podría representar una ventaja logística. No obstante, también es la tienda con menor ingreso total.

La Tienda 3 ofrece un costo de envío relativamente bajo y competitivo, mejor que Tienda 1 y Tienda 2.

## Conclusión y recomendación

Si bien la Tienda 1 lidera en ingresos totales, presenta:

- La calificación promedio más baja

- El costo de envío más alto

Por otro lado, la Tienda 4 tiene el menor costo de envío, pero también el menor ingreso total, lo que puede indicar menor volumen de negocio.

La Tienda 3 se posiciona como la opción más equilibrada porque:

- Tiene ingresos altos y competitivos

- Presenta la mejor calificación promedio (4.05)

- Mantiene altos volúmenes en las categorías más demandadas

- Posee un costo de envío relativamente bajo

En términos estratégicos, la Tienda 3 combina buena rentabilidad, alto volumen de ventas y mayor satisfacción del cliente, factores fundamentales para la sostenibilidad y el crecimiento a largo plazo.

#### A partir del análisis comparativo, la Tienda 4 se posiciona como la menos rentable del conjunto evaluado, debido a:

- El menor ingreso total entre las cuatro tiendas.

- Falta de liderazgo en categorías estratégicas.

- Desempeño promedio en satisfacción del cliente.

- Aunque presenta el menor costo de envío, este factor no logra compensar su menor nivel de facturación general.

## Recomendación final

#### Se recomienda que el Sr. Juan venda la Tienda 4, ya que es la que presenta menor rentabilidad relativa y menor impacto comercial dentro del portafolio analizado.
